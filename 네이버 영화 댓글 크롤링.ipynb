# 네이버 영화 댓글 웹 크롤링


```python
import requests  #http 요청할 때 쓰임
from bs4 import BeautifulSoup   # html에서 정보를 가져오기 위한 라이브러리
```

```python
#기본
test_url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=136990&type=after&onlyActualPointYn=Y&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=2"
resp = requests.get(test_url)
html = BeautifulSoup(resp.content,'html.parser')  # 인스턴스(해당문자열, 파서)
html
```

```python
# <div class:score_result> div:태그명 class:속성 score_result:값

score_result = html.find('div',{'class':'score_result'})
lis = score_result.findAll('li')  #lis 안에 리스트 형식으로 각 댓글의 별점과 댓글 존재
lis[0]
```

```python
## method 로 만들기

def get_data(url):
    resp = requests.get(test_url)
    html = BeautifulSoup(resp.content,'html.parser')
    score_result = html.find('div',{'class':'score_result'})
    lis = score_result.findAll('li') 
    
    watch_movie = []
    
    for li in lis:
        nickname = li.findAll('span')[3].getText()
        review_text = li.find('p').getText()
        score = li.find('em').getText()
        like = li.find('div',{'class':'btn_area'}).findAll('span')[1].getText()
        dislike = li.find('div',{'class':'btn_area'}).findAll('span')[3].getText()
        
        print(nickname,review_text,score,like,dislike)
    
   
```

```python
get_data(test_url)
```

```python
# 전체 댓글수 구하기

result = html.find('div',{'class':'score_total'}).findAll('em')[1].getText()
result = int(result.replace(',',''))

page_num = int(result/10)+1
page_num
```

```python
# 전체 페이지에서 댓글 가져오기
final_url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=136990&type=after&onlyActualPointYn=Y&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false"
resp = requests.get(final_url)
html = BeautifulSoup(resp.content,'html.parser')  # 인스턴스(해당문자열, 파서)

for i in range(1,page_num):
    url = final_url + "&page=" + str(i)
    get_data(url)
```
